```python
import numpy as np
import pandas as pd

df = pd.read_csv('data.csv')
```

```python
df2 = pd.read_csv('file5.csv')
```

```python
filtred_df2 = df2.loc[df2['UCSC_RefGene_Group'].isin({'TSS200','TSS1500'}),['UCSC_RefGene_Name','Name']]
joined_df = pd.merge(filtred_df2, df, left_on='Name', right_on='Unnamed: 0')
final_df = joined_df.groupby(['UCSC_RefGene_Name']).mean()
```

```python
final_df.to_csv('new_data.csv')
```
